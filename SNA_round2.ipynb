```python
import networkx as nx
import random
from tabulate import tabulate
def generate_BA_network(n, m):
    G = nx.barabasi_albert_graph(n, m)
    return G

def assign_activation_probabilities(network):
    activation_probabilities = {}
    for node in network.nodes():
        neighbors = list(network.neighbors(node))
        probs = [random.uniform(0, 1) for _ in range(len(neighbors))]
        sum_probs = sum(probs)
        normalized_probs = [p / sum_probs for p in probs]
        activation_probabilities[node] = {neighbor: prob for neighbor, prob in zip(neighbors, normalized_probs)}
    return activation_probabilities

BA_network = generate_BA_network(1000, 3)
activation_probabilities = assign_activation_probabilities(BA_network)
data = []
random_nodes = random.sample(list(BA_network.nodes()), 10)
for node in random_nodes:
    probs = activation_probabilities[node]
    data.append([node, probs])

print("Randomly selected nodes and their activation probabilities:")
print(tabulate(data, headers=["Node", "Activation Probabilities"], tablefmt="grid"))
```

```python
import networkx as nx
import numpy as np

def run_ICM(network, activation_probabilities):
    activated_nodes = set()
    active_nodes = set()
    steps = 0

    start_node = np.random.choice(list(network.nodes()))
    active_nodes.add(start_node)
    activated_nodes.add(start_node)

    while active_nodes:
        steps += 1
        newly_activated_nodes = set()
        for node in list(active_nodes):
            neighbors = set(network.neighbors(node))
            for neighbor in neighbors:
                if neighbor not in activated_nodes:
                    p = activation_probabilities[node][neighbor]
                    if np.random.rand() < p:
                        newly_activated_nodes.add(neighbor)
                        activated_nodes.add(neighbor)
        active_nodes = newly_activated_nodes

    return steps



steps = run_ICM(BA_network, activation_probabilities)
print("Number of steps required:", steps)

def average_steps_ICM(network, activation_probabilities, num_iterations=10):
    total_steps = 0
    for _ in range(num_iterations):
        steps = run_ICM(network, activation_probabilities)
        total_steps += steps
    average_steps = total_steps / num_iterations
    return average_steps



average_steps = average_steps_ICM(BA_network, activation_probabilities)
print("Average number of steps required (10 iterations):", average_steps)
```

```python
import networkx as nx
import random

def icm(graph, seeds, p=0.1, max_iters=1000):
    active_nodes = set(seeds)
    new_active_nodes = set(seeds)
    max_nodes = len(active_nodes)
    steps = 0

    for i in range(max_iters):
        if len(new_active_nodes) == 0:
            break
        newly_activated = set()
        for node in new_active_nodes:
            neighbors = set(graph.neighbors(node))
            for neighbor in neighbors:
                if neighbor not in active_nodes and random.random() < p:
                    newly_activated.add(neighbor)
        active_nodes |= newly_activated
        new_active_nodes = newly_activated
        if len(active_nodes) > max_nodes:
            max_nodes = len(active_nodes)
            steps = i + 1

    return max_nodes, steps

# Creating sample graphs
G1 = nx.gnm_random_graph(1133, 618, directed=True)
G2 = nx.gnm_random_graph(281903, 2312497, directed=True)

# Running ICM
seeds = [random.choice(list(G1.nodes())) for _ in range(10)]  # Choose 10 random seeds
max_nodes_G1, steps_G1 = icm(G1, seeds)
max_nodes_G2, steps_G2 = icm(G2, seeds)

print("Max  steps on G1:", max_nodes_G1, steps_G1)
print("Max steps on G3:", max_nodes_G2, steps_G2)
```
